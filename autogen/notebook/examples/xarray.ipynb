# XArray
Illustrates how to load multiple results in xarray (multi-demonsional data frames)
For practically, we are using 
the [https://lbrack.testspace.com/](https://lbrack.testspace.com/) organization which doesn't have 
any read restrictions.

```python
from IPython.display import display # Pretty display in Python
import testspace_colab
testspace_colab.use_samples_config() # Comment this line if you want to use you own config.
```

```python
client = testspace_colab.API()
```

```python
import jsonpath_ng
import pandas as pd
from IPython.display import display
import xarray

cfg = dict(project='testspace-samples:getting.started-bitbucket', space='master')

frame_result_details = dict()
# We load all the test cases as pandas DataFrames
for result in client.get_results(**cfg):
    frame_result_details[result['name']] = pd.DataFrame([match.value for match in jsonpath_ng.parse('$..[cases][:]').find( client.get_result_details(result=result['id'], **cfg))])
xr = xarray.Dataset(frame_result_details)
display(xr)
```
